```python
x = np.linspace(0,30,100)
k = 3
rv = sts.chi2(k)
pdf = rv.pdf(x)
plt.plot(x, pdf, label="$k=%s$" % k)
plt.legend()
plt.title("PDF ($\chi^2_k$)")
```

```python
# герерируем хи-квадрат распределение
k = 3 # степени свободы хи-квадрат распределения
s = 1000
rv = sts.chi2(k)
pdf = rv.pdf(x)
plt.plot(x, pdf, label="$k=%s$" % k)
plt.legend()
plt.title("PDF ($\chi^2_k$)")

# выбираем 1000 случайно сгенерированных величин из сгенерированного распределения
plt.hist(rv.rvs(size=s), normed=True, bins=100)
plt.show()
```
