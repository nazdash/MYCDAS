```python
import numpy as np
from matplotlib import pylab as plt
%matplotlib inlineMiscellaneous
```

```python
a = np.arange(-10, 10, 0.1)
```

```python
b = 1 / (1 + np.exp(-a))
plt.plot(a, b)
```

```python
c = np.exp(-a) / (1 + np.exp(-a))
plt.plot(a, c)
```

```python
x = np.arange(0, 1, 0.01)
```

```python
y = np.log(x/(1-x))
plt.plot(x, y)
```

```python
d = np.exp(-a) / (((1 + np.exp(-a))) ** 2)
plt.plot(a, d)
```

```python
e = np.exp(a) / (1 + np.exp(a))
plt.plot(a, e)
```
