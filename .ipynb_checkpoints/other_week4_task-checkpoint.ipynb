```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

```python
lambd = 1
exp_rv = sts.expon(scale = 1/lambd)
# Задаем экспоненциальное распределение с параметром lambda
# Его матожидание равно - 1/lambda, дисперсия - 1/lambda^2
```

```python
sample = exp_rv.rvs(1000)
x = np.linspace(0, 7,100)
# Создаем выборку обьема 1000
```

```python
plt.hist(sample, normed=True, color = 'g')
plt.plot(x, exp_rv.pdf(x),color = 'r')
plt.title('PDF and Histogram')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
# Строим гистограмму по нашей выборке и поверх - ее функцию плотности 
```

```python
n=5
samples = list()
for i in xrange(0,1000):
    samples.append(np.mean(exp_rv.rvs(n)))
mean = 1/lambd
var = (1./(lambd**2))/n
norm_rv = sts.norm(mean, math.sqrt(var))
y = np.linspace(-4,4,100)
plt.hist(samples,normed = 1, color = 'g')
plt.plot(y, norm_rv.pdf(y), 'r-', lw=5, alpha = 0.6)
plt.title('PDF and Histogram(n=5)')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.show()
# Для выборок обьема 5, строим 1000 таких считаем по ним средние. Затем строим гистограмму по средним.
# После, применяем ЦПТ и считаем параметры нормального распределения, которое должно быть у наших средних. 
# Строим функцию распределения поверх, и убеждаемся что они очень близки 
```

```python
 n=10
samples10 = list()
for i in xrange(0,1000):
    samples10.append(np.mean(exp_rv.rvs(n)))
mean = 1/lambd
var = (1./(lambd**2))/n
norm_rv = sts.norm(mean, math.sqrt(var))
y = np.linspace(-4,4,70)
plt.hist(samples10,normed = True, color = 'g')
plt.plot(y, norm_rv.pdf(y), 'r-', lw=5, alpha = 0.6)
plt.title('PDF and Histogram(n=10)')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.show()
# Повторяем действия выше для выборок обьема 10
```

```python
 n=50
samples50 = list()
for i in xrange(0,1000):
    samples50.append(np.mean(exp_rv.rvs(n)))
mean = 1/lambd
var = (1./(lambd**2))/n
norm_rv = sts.norm(mean, math.sqrt(var))
y = np.linspace(-4,4,100)
plt.hist(samples50 ,normed = True, color = 'g')
plt.plot(y, norm_rv.pdf(y), 'r-', lw=5, alpha = 0.6)
plt.title('PDF and Histogram(n=50)')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.show()
# Повторяем те же действия для выборок обьема 50
# Видим, что ЦПТ работает и что с повышением обьема выборок точность аппроксимации повышается
```
